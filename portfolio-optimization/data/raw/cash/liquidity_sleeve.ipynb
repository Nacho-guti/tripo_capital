```python
def tranche_purchase(shortfall, coupon_rate, price_clean):
    """
    Calculate the face value, cost, and coupon income of a bond tranche purchase.
    
    Parameters:
    shortfall (float): Cash needed in that year.
    coupon_rate (float): Annual coupon rate of the bond.
    price_clean (float): Clean price of the bond as a fraction of face value.
    
    Returns:
    tuple: Face value, cost of the bond, and coupon income.
    """
    # Calculate the face value of the bond
    face_value = shortfall / (1 + coupon_rate)
    # Calculate the price of the bond
    price_bond = face_value * price_clean
    # Coupon Income
    coupon_income = face_value * coupon_rate
    # Calculate the face value of the bond
    face_value = shortfall / (1+ coupon_rate)
    # Calculate the price of the bond
    price_bond = face_value * price_clean
    # Coupon Income
    coupon_income = face_value * coupon_rate

    return face_value, price_bond, coupon_income

# Store results for summary table
results = []

# Calculate all tranches
years = [
    (2026, 1, 110_000, 0.015, 97.342 / 100),
    (2027, 2, 110_000, 3.75/100, 99.459 / 100),
    (2028, 3, 110_000, 4.5/100, 101.238 / 100),
    (2029, 4, 110_000, (4 + (1/8))/100, 99.971 / 100),
    (2030, 5, 110_000, (4 + (3/8))/100, 100.731 / 100)
]

# Print summary table header
print("ğŸ¦ BOND LADDER SUMMARY")
print("=" * 100)
print(f"{'Year':<6} {'Yr':<3} {'Coupon':<8} {'Price':<8} {'Face Value':<12} {'Cost Today':<12} {'Coupon Inc':<12} {'Gross Proc':<12}")
print("-" * 100)

total_cost = 0
total_coupon = 0
total_face_value = 0
total_gross = 0
bond_details = []  # Store details for weighted average calculation

for year, yr_num, shortfall, coupon_rate, price_clean in years:
    face_value, price_bond, coupon_income = tranche_purchase(shortfall, coupon_rate, price_clean)
    gross_proceeds = face_value * (1 + coupon_rate)
    
    # Store bond details for later calculations
    bond_details.append((yr_num, price_bond))
    
    # Add to totals
    total_cost += price_bond
    total_coupon += coupon_income
    total_face_value += face_value
    total_gross += gross_proceeds
    
    # Print row
    print(f"{year:<6} {yr_num:<3} {coupon_rate:<7.2%} {price_clean:<7.2%} "
          f"Â£{face_value:<11,.0f} Â£{price_bond:<11,.2f} "
          f"Â£{coupon_income:<11,.0f} Â£{gross_proceeds:<11,.0f}")

# Print totals
print("-" * 100)
print(f"{'TOTAL':<19} Â£{total_face_value:<11,.0f} Â£{total_cost:<11,.2f} "
      f"Â£{total_coupon:<11,.0f} Â£{total_gross:<11,.0f}")
print("=" * 100)

# Calculate return metrics
total_return_amount = total_gross - total_cost
total_return_percentage = (total_return_amount / total_cost) * 100

# Calculate weighted average time to maturity
weighted_time = sum(years_to_maturity * (cost / total_cost) for years_to_maturity, cost in bond_details)

# Calculate annualized return
annualized_return = ((total_gross / total_cost) ** (1/weighted_time) - 1) * 100

# Portfolio summary
print(f"\nğŸ“Š PORTFOLIO SUMMARY:")
print(f"Total Investment Required: Â£{total_cost:,.2f}")
print(f"Total Coupon Income: Â£{total_coupon:,.2f}")
print(f"Total Return at Maturity: Â£{total_gross:,.2f}")
print(f"Net Profit: Â£{total_return_amount:,.2f}")
print(f"Total Return: {total_return_percentage:.2f}%")
print(f"Weighted Average Time to Maturity: {weighted_time:.1f} years")
print(f"Annualized Return: {annualized_return:.2f}%")

```
